```python
#univariate analysis
fig, axis= plt.subplots(2,3,figsize=(10,8)) #2rows,3colms, 6plots at the same time
sns.countplot(ax=axis[0,0],x='Survived', data=df).set_title('Survival Distribution')
sns.countplot(ax=axis[0,1],x='Pclass', data=df).set_title('Passenger Class Distribution')
sns.countplot(ax=axis[0,2],x='Sex', data=df).set_title('Sex Distribution')
sns.countplot(ax=axis[1,0],x='Embarked', data=df).set_title('Embarked Distribution')
sns.countplot(ax=axis[1,1],x='SibSp', data=df).set_title('siblings/spouses Abroad')
sns.countplot(ax=axis[1,2],x='Parch', data=df).set_title('Parents/Children Abroad')

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

```python
plt.figure(figsize=(10,8))      #circles outliers 75-512
sns.boxplot(y='Fare', data=df)
plt.title('Box Plot of Ticket Fare')

plt.ylabel('Fare')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print('\nAnalysing numerical features')

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(16, 6))
fig.suptitle('Univariate Analysis of Numerical Features', fontsize=14)

# Age distribution
sns.histplot(data=df, x='Age', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')

# Fare distribution
sns.histplot(data=df, x='Fare', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Fare Distribution')

# Hide unused subplots
axes[0, 2].axis('off')
axes[1, 0].axis('off')
axes[1, 1].axis('off')
axes[1, 2].axis('off')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python

import seaborn as sns
g=sns.FacetGrid(df, col='Survived', height=4)
g.map(sns.histplot, 'Age', bins=20)
g.fig.suptitle('Age Distribution by Survival ', fontsize=10)
g.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
#new feature
df['FamilySize']=df['SibSp']+df['Parch']+1 #+1 person itself
```

```python
df['IsAlone']=0
df.loc[df['FamilySize']==1, 'IsAlone']=1 #fetch by name
```

```python
df.head()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
sns.barplot(ax=axes[0], x='FamilySize', y='Survived', data=df).set_title('Survival by Family Size')
sns.barplot(ax=axes[1], x='IsAlone', y='Survived', data=df).set_title('Survival rate for travelling alone')
plt.show()
```

```python
df['Title']=df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
print('Extracted Titles:')
df['Title'].value_counts()
```

```python
df['Title']=df['Title'].replace(['Lady','Rev', 'Countess', 'Capt', 'Col','Don','Dr','Major','Sir','Countess','Jonkheer', 'Dona'], 'Rare')
df['Title']=df['Title'].replace('Mlle','Miss')
df['Title']=df['Title'].replace('Ms','Miss')
df['Title']=df['Title'].replace('Mme','Mrs')
plt.figure(figsize=(10,8))
sns.barplot(x='Title', y='Survived', data=df)
plt.title('Survival Plot by title')
plt.ylabel('Survival Probability')
plt.show()

```

```python
sns.catplot(x='Pclass',y='Survived',hue='Sex',data=df,kind='bar', height=6, aspect=1.5)
plt.title('Survival rate by pclass and sex')
plt.ylabel('Survival Probability')
plt.show()
```

```python
import pandas as pd

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df = df.dropna(subset=['Age', 'Sex', 'Survived'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))

sns.violinplot(
    data=df,
    x='Sex',
    y='Age',
    hue='Survived',
    split=True,
    palette={0: 'blue', 1: 'orange'}
)

plt.title('Age Distribution by Sex and Survival')
plt.show()
```

```python
#correlation analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
corr = numeric_df.corr()
```

```python
plt.figure(figsize=(12, 8))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```
